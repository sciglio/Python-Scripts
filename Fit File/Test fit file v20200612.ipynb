### Extracting value from each field and defining a vector

```python
timestamp=[]
position_lat=[]
position_long=[]
distance=[]
altitude=[]
speed=[]
calories=[]
heart_rate=[]
cadence=[]
temperature=[]

with ft.FitReader('data2.fit') as fit:
    for frame in fit:
        if isinstance(frame, ft.FitDataMessage):   
            if frame.name=='record':                
                for j in range(len(frame.fields)):
                    if frame.fields[j].name=='timestamp':
                        timestamp.append(frame.fields[j].value)
                    if frame.fields[j].name=='position_lat':
                        position_lat.append(frame.fields[j].value)
                    if frame.fields[j].name=='position_long':
                        position_long.append(frame.fields[j].value)
                    if frame.fields[j].name=='distance':
                        distance.append(frame.fields[j].value)
                    if frame.fields[j].name=='altitude':
                        altitude.append(frame.fields[j].value)
                    if frame.fields[j].name=='speed':
                        speed.append(frame.fields[j].value)
                    if frame.fields[j].name=='calories':
                        calories.append(frame.fields[j].value)
                    if frame.fields[j].name=='heart_rate':
                        heart_rate.append(frame.fields[j].value)
                    if frame.fields[j].name=='cadence':
                        cadence.append(frame.fields[j].value)
                    if frame.fields[j].name=='temperature':
                        temperature.append(frame.fields[j].value)
```

```python
position_lat=np.array(position_lat)
position_long=np.array(position_long)
distance=np.array(distance)
altitude=np.array(altitude)
speed=np.array(speed)
for i in range(len(speed)):
    if speed[i] is None:
        speed[i]=5
    speed[i]=speed[i]*3.6   
calories=np.array(calories)
heart_rate=np.array(heart_rate)
cadence=np.array(cadence)
temperature=np.array(temperature)
```
