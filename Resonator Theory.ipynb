# Resonators Linewidth, Amplitude and Phase responce 
A small python notebook where the resonators responce is plotted for different parameters and in different envirorment. Initially the resonators are in an matched envirorment; they are studied in **Reflection** near the resonance $f_0$ (only the $S_{11}$ parameter):
$$S_{11}(f)=\frac{\frac{Q_e-Q_i}{Q_e}+2iQ_i\frac{f-f_0}{f}}{\frac{Q_e+Q_i}{Q_e}+2iQ_i\frac{f-f_0}{f}},$$
where $Q_i$ and $Q_e$ are respectively the internal and externa quality factors.


### Import libraries, define function and parameters

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rc('xtick', labelsize=20) 
plt.rc('ytick', labelsize=20) 

def S11(f,f0, Qi, Qe):
    return ((Qe-Qi)/Qe+2*1j*Qi*(f-f0)/f)/((Qe+Qi)/Qe+2*1j*Qi*(f-f0)/f)

def S21_notch(f,f0, Qi, Qe):
    return (Qe+2*1j*Qi*Qe*(f-f0)/f0)/((Qi+Qe+2*1j*Qi*Qe*(f-f0)/f0))


f0=5e9
f=np.linspace(4.99e9, 5.01e9, 1001)
```

### Reflection 

```python
plt.figure(figsize=((10,8)))
plt.title('Reflection resonator IQ plot', fontsize=20)
Qe=10e3
Qi=1e5
plt.plot(S11(f,f0,Qi,Qe).real,S11(f,f0,Qi,Qe).imag, label="Overcoupled")
Qe=10e3
Qi=10e3
plt.plot(S11(f,f0,Qi,Qe).real,S11(f,f0,Qi,Qe).imag, label="Critically coupled")
Qe=1e5
Qi=10e3
plt.plot(S11(f,f0,Qi,Qe).real,S11(f,f0,Qi,Qe).imag, label="Undercoupled")
plt.ylabel('S11 imaginary part', fontsize=20)
plt.xlabel('S11 real part', fontsize=20)
plt.ylim([-1,1])
plt.xlim([-1,1])
plt.legend(fontsize=20)
plt.grid(True)


plt.figure(figsize=((10,8)))
plt.title('Reflection resonator Magnitude plot', fontsize=20)
Qe=1e4
Qi=1e5
plt.plot(f/1e9, 10*np.log10(np.abs(S11(f,f0,Qi,Qe))), label="Overcoupled")
Qe=1e4
Qi=1e4
plt.plot(f/1e9, 10*np.log10(np.abs(S11(f,f0,Qi,Qe))), label="Critically coupled")
Qe=1e5
Qi=1e4
plt.plot(f/1e9, 10*np.log10(np.abs(S11(f,f0,Qi,Qe))), label="Undercoupled")
plt.xlabel('Frequency (GHz)', fontsize=20)
plt.ylabel('S11 magnitude (dB)', fontsize=20)
plt.ylim([-10,1])
plt.legend(fontsize=20)
plt.grid(True)

plt.figure(figsize=((10,8)))
plt.title('Reflection resonator Phase plot', fontsize=20)
Qe=1e4
Qi=1e5
plt.plot(f/1e9, np.angle(S11(f,f0,Qi,Qe)), label="Overcoupled")
Qe=1e4
Qi=1e4
plt.plot(f/1e9, np.angle(S11(f,f0,Qi,Qe)), label="Critically coupled")
Qe=1e5
Qi=1e4
plt.plot(f/1e9,np.angle(S11(f,f0,Qi,Qe)), label="Undercoupled")
plt.xlabel('Frequency (GHz)', fontsize=20)
plt.ylabel('S11 Phase (rad)', fontsize=20)
plt.legend(fontsize=20)
plt.grid(True)
```

### Notch

```python
plt.figure(figsize=((10,8)))
plt.title('Notch resonator IQ plot', fontsize=20)
Qe=10e3
Qi=1e5
plt.plot(S11(f,f0,Qi,Qe).real,S21_notch(f,f0,Qi,Qe).imag, label="Overcoupled")
Qe=10e3
Qi=10e3
plt.plot(S11(f,f0,Qi,Qe).real,S21_notch(f,f0,Qi,Qe).imag, label="Critically coupled")
Qe=1e5
Qi=10e3
plt.plot(S11(f,f0,Qi,Qe).real,S21_notch(f,f0,Qi,Qe).imag, label="Undercoupled")
plt.ylabel('S21 imaginary part', fontsize=20)
plt.xlabel('S21 real part', fontsize=20)
plt.ylim([-1,1])
plt.xlim([-1,1])
plt.legend(fontsize=20)
plt.grid(True)


plt.figure(figsize=((10,8)))
plt.title('Notch resonator Magnitude plot', fontsize=20)
Qe=1e4
Qi=1e5
plt.plot(f/1e9, 10*np.log10(np.abs(S21_notch(f,f0,Qi,Qe))), label="Overcoupled")
Qe=1e4
Qi=1e4
plt.plot(f/1e9, 10*np.log10(np.abs(S21_notch(f,f0,Qi,Qe))), label="Critically coupled")
Qe=1e5
Qi=1e4
plt.plot(f/1e9, 10*np.log10(np.abs(S21_notch(f,f0,Qi,Qe))), label="Undercoupled")
plt.xlabel('Frequency (GHz)', fontsize=20)
plt.ylabel('S21 magnitude (dB)', fontsize=20)
plt.ylim([-10,1])
plt.legend(fontsize=20)
plt.grid(True)

plt.figure(figsize=((10,8)))
plt.title('Notch resonator Phase plot', fontsize=20)
Qe=1e4
Qi=1e5
plt.plot(f/1e9, np.angle(S21_notch(f,f0,Qi,Qe)), label="Overcoupled")
Qe=1e4
Qi=1e4
plt.plot(f/1e9, np.angle(S21_notch(f,f0,Qi,Qe)), label="Critically coupled")
Qe=1e5
Qi=1e4
plt.plot(f/1e9,np.angle(S21_notch(f,f0,Qi,Qe)), label="Undercoupled")
plt.xlabel('Frequency (GHz)', fontsize=20)
plt.ylabel('S21 Phase (rad)', fontsize=20)
plt.legend(fontsize=20)
plt.grid(True)
```
